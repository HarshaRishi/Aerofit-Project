```python
# Representing the marginal probability

marginal_probabilities = round(df['Product'].value_counts(normalize=True) * 100, 2)
print(marginal_probabilities)

# KP281 is the popular choice among the customer, with the highest marginal probability of 44.44%.
# Conversely, KP781 has the lowest marginal probability, indicating that it is purchased less frequently.
```

```python
# Assuming your DataFrame is named 'df'
# Create a cross-tabulation

prod_gender = round(pd.crosstab(df['Product'], df['Gender'], margins=True, margins_name='Total', normalize=True) * 100, 2)
print(prod_gender)

# KP281 is equally popular among females and males.
# Less than 4% Females choose the advanced equipment, KP781.
```

```python
# Assuming your DataFrame is named 'df'
# Create a cross-tabulation
prod_status = round(pd.crosstab(df['Product'], df['MaritalStatus'], margins=True, margins_name='Total', normalize=True) * 100, 2)
print(prod_status)

# KP281 is the popular choice among the customer, with the highest marginal probability of 44.44%.
# Conversely, KP781 has the lowest marginal probability, indicating that it is purchased less frequently.
```

```python
# Assuming your DataFrame is named 'df'
# Create a cross-tabulation
prod_income = round(pd.crosstab(df['Product'], df['Fitness'], margins=True, margins_name='Total', normalize=True) * 100, 2)
print(prod_income)

'''
KP781 is not chosen by customers with fitness less than 3, whereas KP281 and KP481 are the popular choices among the 16% customers who rated 
themselves less than 3.
More than 50% customers rate themselves 3/5.
'''
```

```python
# Fitness is plotted into a count plot for ease of understanding the data.

sns.countplot(data = df, x = 'Fitness')
plt.title('Count plot of Customer Fitness')
plt.show()

# As stated above, more than 50% customers rate themselves 3/5 on Fitness scale. 
```

```python
# Creating kde plot of customer base:

sns.kdeplot(data = df, x = 'Income', color = 'r')
plt.title('KDE Plot of Income of Customer Base')
plt.show()

# Most customers earn between $40000 and $60000.
```

```python
# Creating histplot of customer base:

sns.histplot(data = df, x = 'Miles', color = 'orange')
plt.title('Histogram of Miles expected by the Customer')
plt.show()

# Most customers expect themselves to run 75-125 miles on average.
```

```python
#Creating a heatmap with numeric values in the data like age, education, income, miles, fitness, usage.

# Compute the correlation matrix with all numeric values
corr_map = df.corr(numeric_only = True, method = "pearson")

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_map, annot=True, cmap='flare', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()

# Usage, miles and fitness show a positive correlation with each other.
```

```python
# Creating a Pairplot

sns.pairplot(df, diag_kind='kde')
plt.suptitle('Pair Plot of Variables', y=1.02)
plt.show()
```

```python
# What is the probability of a male customer buying KP781?

# Filter the data for male customers buying KP781
male_kp781 = df[(df['Product'] == 'KP781') & (df['Gender'] == 'Male')]

# Calculate the probability
probability_male_kp781 = len(male_kp781) / len(df)

print(f"Probability of a male customer buying KP781: {probability_male_kp781:.2%}")
```

```python
# What is the probability of a partnered female customer buying KP481?

# Filter the data for partnered female customers buying KP481
female_p_kp481 = df[(df['Product'] == 'KP481') & (df['MaritalStatus'] == 'Partnered') & (df['Gender'] == 'Female')]

# Calculate the probability
probability_female_p_kp481 = len(female_p_kp481) / len(df)

print(f"Probability of a partnered female customer buying KP481: {probability_female_p_kp481:.2%}")
```

```python
# What is the probability of a single female customer buying KP481?

# Filter the data for single female customers buying KP481
female_s_kp481 = df[(df['Product'] == 'KP481') & (df['MaritalStatus'] == 'Single') & (df['Gender'] == 'Female')]

# Calculate the probability
probability_female_s_kp481 = len(female_s_kp481) / len(df)

print(f"Probability of a single female customer buying KP481: {probability_female_s_kp481:.2%}")
```

```python
# What is the probability of a female customer buying KP281?

# Filter the data for female customers buying KP281
female_kp281 = df[(df['Product'] == 'KP281') & (df['Gender'] == 'Female')]

# Calculate the probability
probability_female_kp281 = len(female_kp281) / len(df)

print(f"Probability of a male customer buying KP781: {probability_female_kp281:.2%}")
```

```python
#Conditional Probability of Purchase given Marital Status: Probability of buying a product given the customer is single/partnered.

prob_purchase_given_single = len(df[df['MaritalStatus'] == 'Single']) / len(df)
prob_purchase_given_partnered = len(df[df['MaritalStatus'] == 'Partnered']) / len(df)

print(f"Probability of purchase given single: {prob_purchase_given_single:.2%}")
print(f"Probability of purchase given partnered: {prob_purchase_given_partnered:.2%}")

# Customers who are partnered have a higher probability of approximately 59.44% of making a purchase.
```

```python
# Joint probability: Probability of having high fitness (Fitness >= 4) and high usage (Usage >= 4)

prob_high_fitness_and_usage = len(df[(df['Fitness'] >= 4) & (df['Usage'] >= 4)]) / len(df)
print(f"Probability of high fitness and high usage: {prob_high_fitness_and_usage:.2%}")
```

```python
#Conditional Probability of High Usage given Product: Probability of high product usage (Usage >= 4) given the product is KP781.

prob_high_usage_given_kp781 = len(df[(df['Product'] == 'KP781') & (df['Usage'] >= 4)]) / len(df[df['Product'] == 'KP781'])
print(f"Probability of high usage given KP781: {prob_high_usage_given_kp781:.2%}")

#  97.5% suggests a very high likelihood that customers who purchase the KP781 treadmill are likely to have a high usage level.
```

```python
prob_kp281_given_male = len(df[(df['Product'] == 'KP281') & (df['Gender'] == 'Male')]) / len(df[df['Gender'] == 'Male'])
prob_kp281_given_female = len(df[(df['Product'] == 'KP281') & (df['Gender'] == 'Female')]) / len(df[df['Gender'] == 'Female'])
prob_kp481_given_male = len(df[(df['Product'] == 'KP481') & (df['Gender'] == 'Male')]) / len(df[df['Gender'] == 'Male'])
prob_kp481_given_female = len(df[(df['Product'] == 'KP481') & (df['Gender'] == 'Female')]) / len(df[df['Gender'] == 'Female'])
prob_kp781_given_male = len(df[(df['Product'] == 'KP781') & (df['Gender'] == 'Male')]) / len(df[df['Gender'] == 'Male'])
prob_kp781_given_female = len(df[(df['Product'] == 'KP781') & (df['Gender'] == 'Female')]) / len(df[df['Gender'] == 'Female'])



print(f"Probability of buying KP281 given male: {prob_kp281_given_male:.2%}")
print(f"Probability of buying KP281 given female: {prob_kp281_given_female:.2%}")
print("")
print(f"Probability of buying KP481 given male: {prob_kp481_given_male:.2%}")
print(f"Probability of buying KP481 given female: {prob_kp481_given_female:.2%}")
print("")
print(f"Probability of buying KP781 given male: {prob_kp781_given_male:.2%}")
print(f"Probability of buying KP781 given female: {prob_kp781_given_female:.2%}")

# For KP281, there is a higher probability among females (52.63%) compared to males (38.46%). 
# KP481 also shows a higher probability among females (38.16%) compared to males (29.81%). 
# There is a substantial difference in probabilities for KP781 between males (31.73%) and females (9.21%).
```
